# Analyse Exploratoire (EDA) - CycloPredict
Validation des choix d'ingénierie (Outliers et Lags).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/processed/final_training_set.csv')
df.head()
```

## 1. Détection des Outliers
Vérification des pics de comptage irréalistes.

```python
plt.figure(figsize=(10, 2))
sns.boxplot(x=df['comptage_1h'])
plt.title('Distribution des comptages')
plt.show()
```

## 2. Corrélation des Lags
Validation de la force prédictive des variables de décalage.

```python
cols = ['comptage_1h', 'count_lag_1h', 'count_lag_24h']
sns.heatmap(df[cols].corr(), annot=True, cmap='RdBu_r')
plt.title('Matrice de Corrélation')
plt.show()
```
